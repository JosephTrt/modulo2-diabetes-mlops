# Ingeniería de Características (feature engineering)

```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import joblib
```

```python
# Carga de datos
data_df = pd.read_csv('../data/raw/diabetes_data_upload.csv')
data_df
```

```python
# Quitar espacios en blanco
data_df = data_df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
# Mapeo de caracteristicas

# Caracteristicas si/no
yes_no_map = {'Yes': 1, 'No': 0}
for var in cat_vars[1:-1]:
    data_df[var] = data_df[var].map(yes_no_map)


# Caracteristicas positivo/negativo
class_map = {'Positive': 1, 'Negative': 0}
data_df['class'] = data_df['class'].map(class_map)


# Caracteristicas hombre/mujer
gender_map = {'Male': 1, 'Female': 0}
data_df['Gender '] = data_df['Gender '].map(gender_map)

data_df
```

```python
# Split de los datos
X_train, X_test, y_train, y_test = train_test_split(
    data_df.drop(['class'], axis=1), # predictive variables
    data_df['class'], # target
    test_size=0.2, # portion of dataset to allocate to test set
    random_state=0, # ramdom seed
)

X_train.shape, X_test.shape
```

```python
X_test.head()
```

```python
X_train.head()
```

```python
# Verificacion de valores nulos en X
print(f'X_train -> {[var for var in X_train.columns if X_train[var].isnull().sum() > 0]}')
print(f'X_test -> {[var for var in X_test.columns if X_test[var].isnull().sum() > 0]}')
```

```python
# Escalado
scaler = MinMaxScaler()

# Ajuste
scaler.fit(X_train)

# Recombertir el arry a df
X_train = pd.DataFrame(
    scaler.transform(X_train),
    columns=X_train.columns
)

X_test = pd.DataFrame(
    scaler.transform(X_test),
    columns=X_train.columns
)
```
